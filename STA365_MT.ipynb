## Part 1: Metropolis-Hastings

**Metropolis-Hastings** targetting **stationary distribution** $p$ based on **proposal distribution** $q$ given below

$$q(\tilde x^{(t)} | x^{(t-1)}) = \left(\frac{2}{3}\right)^{1-{\tilde x^{(t)}}}\left(\frac{1}{3}\right)^{\tilde x^{(t)}} \quad
\textrm{ and } \quad p(x^{(t)}) = \left(\frac{1}{3}\right)^{1-{x^{(t)}}}\left(\frac{2}{3}\right)^{x^{(t)}}$$

has **transition kernel** $K$ of transition probabilities 

$K = \left[\begin{array}{cc}\Pr(x^{(t)}=0 \rightarrow x^{(t+1)}=0) & \Pr(x^{(t)}=1 \rightarrow x^{(t+1)}=0)\\\Pr(x^{(t)}=0 \rightarrow x^{(t+1)}=1)& \Pr(x^{(t)}=1 \rightarrow x^{(t+1)}=1)\end{array}\right] = \left[\begin{array}{cl} 2/3 & 1/6 \\ 1/3 & 5/6 \end{array}\right]$


## Part 1A (5%)


### How is this different from `Q3` in `Bayes_03_Metropolis_Hastings.ipynb`<br>and why is this?

In Q3, the proposal distribution was symmetric but here it is asymmetric. This difference in specfication causes the transition kernels to be different. This leads to different acceptance rates and therefore different transition probabilities.

### How is this similar to `Q3` in `Bayes_03_Metropolis_Hastings.ipynb`<br>and why is this true?

The stationary distribution is the same. MH is ensures that regardless of the proposal distribution used, the Markov chain will converge to the target stationary distribution. This works because the acceptance probability is constructed to compensate for any imbalance in the proposal distribution. 



## Part 1B (5%)

### What is the Metropolis-Hastings acceptance probability $\alpha_{mh}$ here?

$\alpha_{MH}(\tilde{x}|x) = \min\left(1, \frac{p(\tilde{x})q(x|\tilde{x})}{p(x)q(\tilde{x}|x)}\right)$

Let's calculate for both possible transitions:

1. For $x^{(t-1)} = 0 \rightarrow \tilde{x}^{(t)} = 1$:

$\alpha_{MH}(1|0) = \min\left(1, \frac{p(1)q(0|1)}{p(0)q(1|0)}\right) = \min\left(1, \frac{(2/3)(2/3)}{(1/3)(1/3)}\right) = \min(1, 4) = 1$

2. For $x^{(t-1)} = 1 \rightarrow \tilde{x}^{(t)} = 0$:

$\alpha_{MH}(0|1) = \min\left(1, \frac{p(0)q(1|0)}{p(1)q(0|1)}\right) = \min\left(1, \frac{(1/3)(1/3)}{(2/3)(2/3)}\right) = \min(1, 1/4) = 1/4$

Therefore, the acceptance probabilities are:

 - When proposing a move from state 0 to state 1: accept with probability 1

 - When proposing a move from state 1 to state 0: accept with probability 1/4

### What is the difference between the *acceptance probability,  proposal distribution, transition kernel, stationary distribution and target distribution*?

The proposal distribution generates candidate states for the next step in the Markov chain. The acceptance probability then determines where to accept this proposal based on the ratio of target and proposal densities. Altogether this process results in the transition kernel - a probability matrix that defines state to state moves, balancing exploration with convergence to the target. The stationary distribution is achieved regardless of the proposal mechanism used, which are the probabilities that remain unchanged after applying the transition kernel and match our target distribution ($p(x) = (1/3)^{1-x}(2/3)^x$).

### Use the probabilities from the table above

$\begin{align}
\Pr(s_0 \rightarrow s_0) &= q(0|0) + q(1|0)(1-\alpha_{0 \rightarrow 1}) \\
&= 2/3 + (1/3)(1-1) \\
&= 2/3 + 0 \\
&= 2/3
\end{align}$
$\begin{align}
\Pr(s_0 \rightarrow s_1) &= q(1|0)\alpha_{0 \rightarrow 1} \\
&= (1/3)(1) \\
&= 1/3
\end{align}$
$\begin{align}
\Pr(s_1 \rightarrow s_0) &= q(0|1)\alpha_{1 \rightarrow 0} \\
&= (2/3)(1/4) \\
&= 1/6
\end{align}$
$\begin{align}
\Pr(s_1 \rightarrow s_1) &= q(1|1) + q(0|1)(1-\alpha_{1 \rightarrow 0}) \\
&= 1/3 + (2/3)(1-1/4) \\
&= 1/3 + (2/3)(3/4) \\
&= 1/3 + 1/2 \\
&= 5/6
\end{align}$
Therefore, the transition kernel is:
$K = \left[\begin{array}{cc}
\Pr(s_0 \rightarrow s_0) & \Pr(s_1 \rightarrow s_0)\\
\Pr(s_0 \rightarrow s_1) & \Pr(s_1 \rightarrow s_1)
\end{array}\right] = \left[\begin{array}{cl}
2/3 & 1/6 \\
1/3 & 5/6
\end{array}\right]$

## Part 2 (20%): Complete the specification above using PyMC
- Use one beta distribution
- Use one Bernoulli distribution with `shape=272`
- Use four normal distributions for the location $\theta$ parameters, each with `shape=1`
- Use two gamma distributions for the precision $\tau$ parameters, each with `shape=1`
- Use two normal distributions for the two observation columns (as given below)

```python
# If u is a vector (with `shape=272` in this case)
# then `(1-u)*theta_w0 + u*theta_w1` creates a vector of length 272
# which can be used for the mean of the observed `w` data along with `tau=tau_w`
# so long as `shape=1` is used in the specification of `theta_w0` and `theta_w1`
# which makes these random variables a "vector" as opposed to a "scalar"

with pymc.Model() as OldFaithful:
    # Beta prior for p
    p = pymc.Beta("p", alpha=1, beta=1)

    #bernoulli for cluster assignment
    u = pymc.Bernoulli("u", p=p, shape=272)

    #normal distributions for waiting durations
    theta_w0 = pymc.Normal("theta_w0", mu=55, tau=1/5**2, shape=1)
    theta_w1 = pymc.Normal("theta_w1", mu=85, tau=1/5**2, shape=1)

    #normal distributions for eruption durations
    theta_y0 = pymc.Normal("theta_y0", mu=2.0, tau=1/0.5**2, shape=1)
    theta_y1 = pymc.Normal("theta_y1", mu=4.5, tau=1/0.5**2, shape=1)

    #gamma distribution for precision parameters
    tau_w = pymc.Gamma("tau_w", alpha=1/2, beta=1/2*10**2)
    tau_y = pymc.Gamma("tau_y", alpha=1/2, beta=1/2*1**2)
    # The observed data flows into the model through two entry points
    
    obs_y = pymc.Normal("obs_y", mu = (1-u)*theta_y0 + u*theta_y1, 
                        tau=tau_y, observed=old_faithful.eruptions)
    obs_w = pymc.Normal("obs_w", mu = (1-u)*theta_w0 + u*theta_w1, 
                        tau=tau_w, observed=old_faithful.waiting)
    
```

## Part 4D (10%): Complete Gibbs Sampler


```python
n = 272
G = 100

thetas_w_mu0 = np.array([55,85])
thetas_w = np.zeros((2,G)) + thetas_w_mu0.reshape(2,1)
thetas_w_tau0 = 1/5**2

thetas_y_mu0 = np.array([2.0,4.5])
thetas_y = np.zeros((2,G)) + thetas_y_mu0.reshape(2,1)
thetas_y_tau0 = 1/0.5**2

taus_alpha = 1/2
taus_beta = np.array([1/2 * 10**2, 1/2 * 1**2])
taus = np.zeros((2,G)) + taus_alpha/taus_beta.reshape(2,1)

us = np.zeros((n,G))
ps = np.zeros(G)+0.5
waiting = old_faithful.waiting.values
eruptions = old_faithful.eruptions.values
for g in range(1,G):
    #current parameters
    tau_w = taus[0, g-1]
    tau_y = taus[1, g-1]
    theta_w0 = thetas_w[0, g-1]
    theta_w1 = thetas_w[1, g-1]
    theta_y0 = thetas_y[0, g-1]
    theta_y1 = thetas_y[1, g-1]
    p = ps[g-1]

    #probabilities for cluster assignments
    density_w0 = stats.norm.pdf(waiting, loc=theta_w0, scale=1/np.sqrt(tau_w))
    density_w1 = stats.norm.pdf(waiting, loc=theta_w1, scale=1/np.sqrt(tau_w))
    density_y0 = stats.norm.pdf(eruptions, loc=theta_y0, scale=1/np.sqrt(tau_y))
    density_y1 = stats.norm.pdf(eruptions, loc=theta_y1, scale=1/np.sqrt(tau_y))
    
    #calculate posterior probabilities
    prob_u1 = density_w1 * density_y1 * p
    prob_u0 = density_w0 * density_y0 * (1-p)
    prob_u = prob_u1 / (prob_u1 + prob_u0)

    #sample new cluster assignments
    us[:,g] = stats.bernoulli.rvs(prob_u)
    alpha_post = 1 + np.sum(us[:,g]==1)
    beta_post = 1 + np.sum(us[:,g]==0)
    ps[g] = stats.beta.rvs(alpha_post, beta_post)

    # update theta params
    w0_data = waiting[us[:,g]==0]
    w1_data = waiting[us[:,g]==1]
    y0_data = eruptions[us[:,g]==0]
    y1_data = eruptions[us[:,g]==1]

    n0 = len(w0_data)
    n1 = len(w1_data)

    prec = thetas_w_tau0 + n0*tau_w
    tmp = thetas_w_tau0 * thetas_w_mu0[0] + tau_w * np.sum(w0_data)
    mu = tmp/prec
    thetas_w[0,g] = stats.norm.rvs(loc=mu, scale=1/np.sqrt(prec))

    prec = thetas_w_tau0 + n1*tau_w
    tmp = thetas_w_tau0 * thetas_w_mu0[1] + tau_w * np.sum(w1_data)
    mu = tmp/prec
    thetas_w[1,g] = stats.norm.rvs(loc=mu, scale=1/np.sqrt(prec))
    prec = thetas_y_tau0 + n0*tau_y
    tmp = thetas_y_tau0 * thetas_y_mu0[0] + tau_y * np.sum(y0_data)
    mu = tmp/prec
    
    thetas_y[0,g] = stats.norm.rvs(loc=mu, scale=1/np.sqrt(prec))
    prec = thetas_y_tau0 + n1*tau_y
    tmp = thetas_y_tau0 * thetas_y_mu0[1] + tau_y * np.sum(y1_data)
    mu = tmp/prec
    thetas_y[1,g] = stats.norm.rvs(loc=mu, scale=1/np.sqrt(prec))

    #update tau 
    w0_data = waiting[us[:,g]==0]
    w1_data = waiting[us[:,g]==1]
    y0_data = eruptions[us[:,g]==0]
    y1_data = eruptions[us[:,g]==1]

    sum_sq_w0 = np.sum((w0_data - thetas_w[0,g])**2)
    sum_sq_w1 = np.sum((w1_data - thetas_w[1,g])**2)
    total_sq_w = sum_sq_w0 + sum_sq_w1
    
    alpha_new_w = taus_alpha + (n0+n1)/2
    beta_new_w = taus_beta[0] + total_sq_w/2
    taus[0,g] = stats.gamma.rvs(alpha_new_w, scale=1/beta_new_w)

    sum_sq_y0 = np.sum((y0_data - thetas_y[0,g])**2)
    sum_sq_y1 = np.sum((y1_data - thetas_y[1,g])**2)
    total_sq_y = sum_sq_y0 + sum_sq_y1
    alpha_new_y = taus_alpha + (n0+n1)/2
    beta_new_y = taus_beta[1] + total_sq_y/2
    taus[1,g] = stats.gamma.rvs(alpha_new_y, scale=1/beta_new_y)
```
