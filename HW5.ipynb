In the first model we use a standard conjugate prior formulation:
Prior for $\theta$:  $\theta \sim \mathcal{N}(5, 10^2)$ .
(A weakly informative normal prior centered at 5.)
Prior for $\tau$:  $\tau \sim \text{Gamma}(\alpha=2, \beta=1)$ .
(A gamma prior on the precision parameter.)

The likelihood is as given above. Notice that we “translate” from the precision $\tau$ to the standard deviation for the normal likelihood by setting

$\sigma = \frac{1}{\sqrt{\tau}}$.


For Model 2 we change both priors:
Prior for $\theta$: Use a Laplace (double exponential) distribution

$p(\theta) \propto \exp\Bigl(-\frac{|\theta-5|}{1.0}\Bigr)$

Prior for $\tau$: Use a lognormal distribution

$\tau \sim \text{LogNormal}(\mu=-1, \sigma=1)$.


In the third model we choose:
Prior for $\theta$: A Student’s t distribution with 3 degrees of freedom

$\theta \sim \text{StudentT}(\nu=3, \mu=5, \sigma=2)$

(A heavy-tailed prior which can be more robust.)
	- Prior for $\tau$: A Uniform distribution on a positive interval, e.g., $\tau \sim \text{Uniform}(0.001, 10)$.
(This is a non‐gamma prior that lets $\tau$ be anywhere in a plausible range.)



```python
import numpy as np
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt

# For reproducibility
np.random.seed(42)

# Generate synthetic data: mean 5 and standard deviation 2.
x = np.random.normal(5, 2, 100)
with pm.Model() as model1:
    # Prior for theta: Normal
    theta = pm.Normal('theta', mu=5, sigma=10)
    # Prior for tau: Gamma (shape-rate parameterization)
    tau = pm.Gamma('tau', alpha=2, beta=1)
    # Deterministic transformation: sigma = 1/sqrt(tau)
    sigma = pm.Deterministic('sigma', 1/pm.math.sqrt(tau))
    
    # Likelihood: note that the observed data x are modeled as normal with mean theta and std sigma.
    x_obs = pm.Normal('x_obs', mu=theta, sigma=sigma, observed=x)
    
    # Draw samples using NUTS (the default sampler)
    trace1 = pm.sample(2000, tune=1000, target_accept=0.9, return_inferencedata=True)
    
# Diagnostics for Model 1
print("Model 1 Summary:")
print(az.summary(trace1, round_to=2))
az.plot_trace(trace1)
plt.show()

with pm.Model() as model2:
    # Prior for theta: Laplace (double exponential)
    theta = pm.Laplace('theta', mu=5, b=1.0)
    # Prior for tau: LogNormal (a non-Gamma alternative)
    tau = pm.Lognormal('tau', mu=-1, sigma=1)
    sigma = pm.Deterministic('sigma', 1/pm.math.sqrt(tau))
    
    x_obs = pm.Normal('x_obs', mu=theta, sigma=sigma, observed=x)
    
    trace2 = pm.sample(2000, tune=1000, target_accept=0.9, return_inferencedata=True)
    
# Diagnostics for Model 2
print("Model 2 Summary:")
print(az.summary(trace2, round_to=2))
az.plot_trace(trace2)
plt.show()

with pm.Model() as model3:
    # Prior for theta: Student's t with 3 degrees of freedom (heavy-tailed)
    theta = pm.StudentT('theta', nu=3, mu=5, sigma=2)
    # Prior for tau: Uniform over a plausible range of precision values
    tau = pm.Uniform('tau', lower=0.001, upper=10)
    sigma = pm.Deterministic('sigma', 1/pm.math.sqrt(tau))
    
    x_obs = pm.Normal('x_obs', mu=theta, sigma=sigma, observed=x)
    
    trace3 = pm.sample(2000, tune=1000, target_accept=0.9, return_inferencedata=True)
    
# Diagnostics for Model 3
print("Model 3 Summary:")
print(az.summary(trace3, round_to=2))
az.plot_trace(trace3)
plt.show()
```

# Q3
Slice sampling augments the target distribution  f(x)  (known only up to a constant) by introducing an auxiliary variable  y . For a given current value  x , we:
 - Draw a vertical level  y  uniformly between 0 and  f(x) . This gives a horizontal “slice” under the curve: $S = \{ x: f(x) > y \}$.

- Then sample a new $x$ uniformly from the slice $S$.


In the code, the function `slice_f_at_y` starts with a grid of  x -values (by default, `x_grid = np.linspace(0,1,51)`) and identifies those grid points where  $f(x) > y$ . This determines an initial interval $[a, b]$ that covers the slice. The endpoints are then extended by one grid step so that at the endpoints $f(a) < y$ and $f(b) < y$.

A candidate  $x^*$  is drawn uniformly from the interval [a, b].
 - If  $f(x^*) > y$  (i.e., the candidate lies “under” the curve  $f$  at level  $y$ ), it is immediately accepted.
 - Otherwise, the candidate is rejected and the interval is shrunk. The direction of the shrinkage depends on whether  $x^*$  lies to the left or right of the current  $x$ .
 - The helper function `slice_f_at_y_` recursively refines the interval until a candidate is accepted.

When the full conditional distribution for a parameter is known only up to a normalizing constant, standard Gibbs sampling (which requires exact draws) becomes challenging. Typically, an MH step might be used, but slice sampling offers an alternative that does not require proposal tuning.
